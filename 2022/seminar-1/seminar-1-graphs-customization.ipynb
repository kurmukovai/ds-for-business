# 5* standard types of graphs and how to plot them

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Index chart (or Line chart)

```python
plt.plot([5,10,20, 30], [1, 2, 3, 4])
```

### Order matters

```python
plt.plot([10, 5, 30, 20], [2, 1, 4, 3])
```

### Labels

```python
plt.plot([5,10,20, 30], [1, 2, 3, 4])
plt.ylabel('some numbers')
plt.xlabel('other numbers');
```

### Increase font size (global)

```python
font = {'size'   : 18}
plt.rc('font', **font)
```

### Color

https://matplotlib.org/2.0.2/examples/color/named_colors.html

```python
plt.plot([5,10,20, 30], [1, 2, 3, 4], color='k')
plt.ylabel('some numbers')
plt.xlabel('other numbers');
```

You could use hex color encoding (for example from here https://coolors.co)

```python
plt.plot([5,10,20, 30], [1, 2, 3, 4], color='#FE4A49')
plt.ylabel('some numbers')
plt.xlabel('other numbers');
```

### Line style

https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html

```python
plt.plot([5,10,20, 30], [1, 2, 3, 4],
         color = 'darkmagenta',
         linestyle='-.')

plt.ylabel('some numbers')
plt.xlabel('other numbers');
```

### Title

```python
plt.plot([5,10,20, 30], [1, 2, 3, 4],
         color = 'darkmagenta',
         linestyle='--')

plt.title('Some title')
plt.ylabel('some numbers')
plt.xlabel('other numbers');
```

### Multiple lines

```python
plt.plot([5,10,20, 30], [1, 2, 3, 4],
         color = 'darkmagenta',
         linestyle='--')

plt.plot([1,2,3, 4,], [1, 4, 9, 16],
         color = 'red',
         linestyle='dotted')

plt.title('Some title')
plt.ylabel('some numbers')
plt.xlabel('other numbers');
```

### Legend

```python
plt.plot([5,10,20, 30], [1, 2, 3, 4],
         color = 'darkmagenta',
         linestyle='--',
         label='One line')

plt.plot([1,2,3, 4,], [1, 4, 9, 16],
         color = 'red',
         linestyle='dotted',
         label='Another line')

plt.legend()
plt.title('Some title')
plt.ylabel('some numbers')
plt.xlabel('other numbers');
```

# Bar chart

```python
x = [1, 2, 3]
sizes = [150, 93, 40]

plt.bar(x, sizes);
```

### Labels

```python
x = [1, 2, 3]
sizes = [150, 93, 40]
labels = ['vegetables', 'fruits', 'meat']

plt.bar([1,2,3], sizes, tick_label=labels);
```

### Color

```python
x = [1, 2, 3]
sizes = [150, 93, 40]
labels = ['vegetables', 'fruits', 'meat']
c = ['green', 'orange', 'red']

plt.bar([1,2,3], sizes,tick_label=labels, color=c);
```

### Width

```python
x = [1, 2, 3]
sizes = [150, 93, 40]
labels = ['vegetables', 'fruits', 'meat']
c = ['green', 'orange', 'red']

plt.bar([1,2,3], sizes, width=0.5, tick_label=labels, color=c);
```

### Horizontal

```python
x = [1, 2, 3]
sizes = [150, 93, 40]
labels = ['vegetables', 'fruits', 'meat']
c = ['green', 'orange', 'red']

plt.barh([1,2,3], sizes, height=0.5, tick_label=labels, color=c);
```

### Add grid, fix xticks

```python
x = [1, 2, 3]
sizes = [150, 93, 40]
labels = ['vegetables', 'fruits', 'meat']
c = ['green', 'orange', 'red']

plt.grid(axis='x')
plt.xticks([0, 30, 60, 90, 120, 150])
plt.barh([1,2,3], sizes, height=0.5, tick_label=labels, color=c);
```

# Piechart*

> I personally do not like piecharts. You should always use barchart instead.
> But in case you want to use them anyway, you better know some options.

> Piecharts are the worst https://www.data-to-viz.com/caveat/pie.html

```python
sizes = [257, 287] # ARC, AREA of a segment, ANGLE

plt.pie(sizes);
```

### Labels

```python
sex = ['Males', 'Females']


plt.pie(sizes, labels=sex);
```

### Start angle

```python
plt.pie(sizes, labels=sex, startangle=90);
```

### Numeric value on pie pieces

```python
plt.pie(sizes, labels=sex, startangle=90, autopct=f'%.1f');
```

```python
def num_label(percentage):
    return f'{int(percentage / 100 * 544)}'
```

```python
plt.pie(sizes, labels=sex, startangle=90, autopct=num_label);
```

### Explode

```python
explode = [0.25, 0]

plt.pie(sizes,
        labels=sex, 
        startangle=90,
        autopct=num_label,
        explode=explode);
```

```python
sizes = [150, 93, 40]
labels = ['vegetables', 'fruits', 'meat']

explode = [0.05, 0, 0]
plt.pie(sizes,
        labels=labels, 
        startangle=90,
        autopct=num_label,
        explode=explode);
```

```python
def num_label(percentage):
    return f'{int(round(percentage))}%'
```

```python
explode = [0.05, 0, 0]
plt.pie(sizes,
        labels=labels, 
        startangle=90,
        autopct=num_label,
        explode=explode);
```

### Pie size (radius)

```python
sizes = [150, 93, 40]

labels = ['vegetables', 'fruits', 'meat']
explode = [0.05, 0, 0]
plt.pie(sizes,
        labels=labels, 
        startangle=90,
        autopct=num_label,
        radius=2);
```

### Pie chart wedge properties (Donut chart)

```python
sizes = [150, 93, 40]
labels = ['vegetables', 'fruits', 'meat']

width = 1
explode = [0.05, 0, 0]
plt.pie(sizes,
        labels=labels, 
        startangle=90,
        autopct=num_label,
        radius= 2,
        wedgeprops={'width': width});
```

### Title

```python
sizes = [150, 93, 40]
labels = ['vegetables', 'fruits', 'meat']

width = .6
explode = [0.05, 0, 0]
plt.pie(sizes,
        labels=labels, 
        startangle=90,
        autopct=num_label,
        radius= 1,
        wedgeprops={'width': width});

plt.title('Food busket');
```

### seaborn's boxplots are much prettier
https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
import seaborn as sns
```

```python
df['Gender'] = df.male.apply(lambda x: 'Male' if x == 1 else 'Female')
```

```python
sns.boxplot(y='height', data = df, width=0.5);
```

```python
sns.boxplot(x = 'Gender', y='height', data = df, width=0.1);
```

### Orientation

```python
sns.boxplot(x = 'height', y='Gender', data = df, width=0.5, orient='h');
```

### Compare with histogram

```python
plt.hist([df.height[df.male==1], df.height[df.male==0]],
         bins = 10,
        label = ['Males', 'Females']);

plt.legend()
plt.ylabel('#number of people')
plt.xlabel('Height, cm');
```

### groupped

```python
df
```

```python
def get_age_group(age):
    if age < 3:
        return 'baby'
    elif age < 13:
        return 'child'
    elif age < 21:
        return 'teenager'
    elif age < 50:
        return 'adult'
    else:
        return 'senior'
        
df['Age group'] = df.age.apply(get_age_group)
```

```python
sns.boxplot(x = 'Gender', y='height', hue = 'Age group', data = df);
```

### increase figure size

```python
fig, ax = plt.subplots(figsize=(12, 7))
sns.boxplot(x = 'Gender', y='height', hue = 'Age group', data = df, ax=ax);
```
